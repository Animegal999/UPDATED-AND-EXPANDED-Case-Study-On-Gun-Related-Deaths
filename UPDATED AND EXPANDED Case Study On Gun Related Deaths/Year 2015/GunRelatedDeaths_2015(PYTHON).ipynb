There are 15 csv files inside the kaggle source, which have been separated and sorted into respective files, as well as the Jupyter Notebooks for each file. This file is for the year 2014 and any gun related deaths that took place during that time.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
from scipy import stats

%matplotlib inline
```

```python
# Loading in the csv file for the year of 2015:
gunDeaths_2015_df = pd.read_csv("guns_2015.csv")
gunDeaths_2015_df.head()
```

```python
# Double checking for nulls
pd.isna(gunDeaths_2015_df)
```

```python
# Describing the age column, as well as which month has the most gun related deaths
gunDeaths_2015_df['age'].describe()
```

```python
print(statistics.mode(gunDeaths_2015_df['month']))
# The month of July had the most gun related deaths in 2015.
```

```python
print(statistics.mode(gunDeaths_2015_df['age']))
# 24 year olds were involved with gun related incidents, the most often, in 2015.
```

```python
# Mode can also deduce from categorical statistics.
# Which race was involved the most with these incidents?
print(statistics.mode(gunDeaths_2015_df['race']))
# White/Caucasian people were involved the most with gun related deaths in 2015.
```

```python
# Where were people killed the most, if a gun was involved?
print(statistics.mode(gunDeaths_2015_df['place']))
# Most victims were killed at their homes, in 2015.
```

```python
# What was the highest level of education for these victims?
print(statistics.mode(gunDeaths_2015_df['education']))
# Most victims had either their High School Diploma, or GED.
```

```python
# Were most of the victims male or female?
print(statistics.mode(gunDeaths_2015_df['sex']))
# Most of the victims were males.
```

```python
# Are there any outliers inside the age column?
sns.boxplot(x=gunDeaths_2015_df['age'])
# No outliers as well
```

```python
# Plotting a histogram for the gun-related deaths (this confirms the mode to be July/7)
gunDeaths_2015_df['month'].hist(bins=24)
plt.title('Frequency of Gun Related Deaths by Month')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting a histogram for the victims' race 
gunDeaths_2015_df['race'].hist(bins=8)
plt.title('Frequency of Gun Related Deaths by Race')
plt.xlabel('Race')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting a Histogram for gun related deaths in 2015, by gender:
# (this confirms the mode to be Males/M)
gunDeaths_2015_df['sex'].hist(bins=3)
plt.title('Frequency of Gun Related Deaths by Gender')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting a Histogram for gun related deaths in 2015, by where the victim was killed:
# (this confirms the mode to be at home)
gunDeaths_2015_df['place'].hist(bins=20)
plt.title('Frequency of Gun Related Deaths by Location')
plt.xlabel('Location')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting a histogram for the intentions of using a gun. This confirms the most common intention to be Suicide.
gunDeaths_2015_df['intent'].hist(bins=10)
plt.title('Frequency of Gun Related Deaths by Intentions')
plt.xlabel('Intention(s)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting a histogram for the hightest level of education of the victim.
# This confirms the most common level is High School Diploma/GED
gunDeaths_2015_df['education'].hist(bins=10)
plt.title('Frequency of Gun Related Deaths by Victim\'s Education Level')
plt.xlabel('Education Level')
plt.ylabel('Frequency')
plt.show()
```

```python
# Were the police contacted often for these incidents?
gunDeaths_2015_df['police'].hist(bins=3)
plt.title('Frequency of Police Intervention')
plt.xlabel('0 (no), 1 (yes)')
plt.ylabel('Frequency')
plt.show()
# The police were not contacted for a majority of these incidents
```

Conclusive analysis: Most of the victims from this dataset either:

- Were male
- Were 24 years old
- Had either their High School Diploma, or their GED
- Died to a Suicide attempt
- Died at home
- Died in the month of July

Comparing to the previous year:

- Mode age is higher, but the mean age of the victims was slightly lower in 2015.
- More people died overall in 2015.
- More people died in July of 2015 than in August of 2014.
