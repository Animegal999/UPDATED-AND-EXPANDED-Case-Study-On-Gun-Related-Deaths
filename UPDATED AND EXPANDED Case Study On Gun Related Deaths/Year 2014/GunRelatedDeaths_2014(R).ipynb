```r
install.packages("ggplot2")
install.packages("tinytex")
install.packages("rmarkdown", dep = TRUE)
install.packages("plotly")
install.packages("DescTools")
install.packages("dplyr")

library(ggplot2)
library(tinytex)
library(rmarkdown)
library(plotly)
library(DescTools)
library(dplyr)
```

```r
gunDeathsData_2014 <- read.csv("/content/guns_2014.csv", header = TRUE, sep = ",")
View(gunDeathsData_2014)
```

```r
# Any null data?
sum(is.na(gunDeathsData_2014))
```

```r
# Gun related deaths from 2014
# Statistical information

# Mean for each numeric column:
# Only one numeric column was found: Age
print("Mean values for each numeric column:")
print("Mean Age: ") # 44.228
mean(gunDeathsData_2014$age)
print("-----------------------------")
```

```r
# Median Age:
print("Median value for the age column:") # 42
median(gunDeathsData_2014$age)
```

```r
# Mode Age:
print("Most occuring value for the age column:") # 22
Mode(gunDeathsData_2014$age)
```

```r
# Standard Deviation Age:
print("Standard Deviation of the age:") #19.699 years
SD(gunDeathsData_2014$age)
```

```r
# creating boxplot on age and race:
boxplot(gunDeathsData_2014$age ~ gunDeathsData_2014$race, data = gunDeathsData_2014,
        main = "Age of Victims of Gun-related Deaths by Race",
        xlab = "Race",
        ylab = "Age")
```

```r
# Constructing the histogram:

hist(gunDeathsData_2014$age, xlab="Victims of Gun Violence By Age", col = "red", border = "black")
```

```r
# Creating a boxplot on age and gender:
boxplot(gunDeathsData_2014$age ~ gunDeathsData_2014$sex, data = gunDeathsData_2014,
        main = "Age of Victims of Gun-related Deaths by Gender",
        xlab = "Gender",
        ylab = "Age")
```

```r
# Constructing the histogram for number of deaths by month:

hist(gunDeathsData_2014$month, xlab="Total Victims of Gun Violence By Month", col = "blue", border = "black")
```

```r
# Creating a boxplot on age and intentions behind the gun:
boxplot(gunDeathsData_2014$age ~ gunDeathsData_2014$intent, data = gunDeathsData_2014,
  main = "Age of victims of Gun-related Deaths by Intent",
  xlab = "Intentions With the Gun",
  ylab = "Age")

# Homicides have the most outliers
```
